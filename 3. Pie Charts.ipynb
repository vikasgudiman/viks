```python
import pandas as pd
import numpy as np
df=pd.read_csv("s1.csv")
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
df2=df[["Respondent","LanguageWorkedWith"]]
unique_languages={}
test=[]
for langs in df2['LanguageWorkedWith'].values:
    if langs is np.nan:
        continue
    else:
        for lang in langs.split(";"):
            if lang in unique_languages:
                unique_languages[lang]+=1
            else:
                unique_languages[lang]=1
unique_languages = dict(sorted(unique_languages.items(), key=lambda kv: kv[1],reverse=True)[0:5])
x=list(unique_languages.keys())
y=list(unique_languages.values())
```

```python
#Pie charts only for data < 5 else bar
plt.pie(y,labels=x)
plt.title('Languages popularity')#Title of the graph
plt.tight_layout()
plt.show()
```

```python
#explode values
explode = [0,0,0,0.3,0]#Change the value to explode mode. It is a ratio of radius
plt.pie(y,labels=x,explode=explode)
plt.title('Languages popularity')#Title of the graph
plt.tight_layout()
plt.show()
```

```python
#3d look shadow=True
explode = [0,0,0,0.1,0.2]#Change the value to explode mode. It is a ratio of radius
plt.pie(y,labels=x,explode=explode,shadow=True)
plt.title('Languages popularity')#Title of the graph
plt.tight_layout()
plt.show()
```

```python
#display percentages in pie charts
explode = [0,0,0,0.1,0.2]#Change the value to explode mode. It is a ratio of radius
plt.pie(y,labels=x,explode=explode,shadow=True,autopct="%1.1f%%")
plt.title('Languages popularity')#Title of the graph
plt.tight_layout()
plt.show()
```
