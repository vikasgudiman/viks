```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.style.use("fivethirtyeight")
```

```python
ages=np.array([5,6,10,20,21,22,23,45,46,77,99,100])
ages
```

```python
plt.title("Age of Respondants")
plt.xlabel("Age")
plt.ylabel("Respondants")
bins=[0,20,40,60,80,100]
plt.hist(ages,bins=bins,edgecolor="black")
plt.tight_layout()
```

```python
import pandas as pd
```

```python
df=pd.read_csv("s1.csv")
```

```python
x=df["Respondent"].values
y=df["Age"].values
median_val=df["Age"].median()
```

```python
plt.title("Age of Respondants")
plt.xlabel("Age")
plt.ylabel("Respondants")
bins=[0,20,40,60,80,100]
plt.hist(y,bins=bins,edgecolor="black")
plt.tight_layout()
```

```python
plt.title("Age of Respondants")
plt.xlabel("Age")
plt.ylabel("Respondants")
bins=[0,20,40,60,80,100]
plt.hist(y,bins=bins,edgecolor="black",log=True)
plt.tight_layout()
```

```python
plt.title("Age of Respondants")
plt.xlabel("Age")
plt.ylabel("Respondants")
bins=[0,20,40,60,80,100]
plt.hist(y,bins=bins,edgecolor="black",log=True)
plt.axvline(median_val,color="#fc4f30",label="Median Age")
plt.legend()
plt.tight_layout()
```
